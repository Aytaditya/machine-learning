```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

data={
    'Name':['Kristina', 'nick', 'krish', 'Rose'],
    "City":['New York', 'Los Angeles', 'New York', 'Chicago'],
    "Test Passed":['Yes', 'No', 'Yes', 'No'],
    "Gender":['Female','Male','Male','Female']
}

df=pd.DataFrame(data)
df
```

# LABEL ENCODING

```python
df_label=df.copy()  
le=LabelEncoder()

df_label['Gender Encoded']=le.fit_transform(df_label['Gender']) 
df_label['Passed Encoded']=le.fit_transform(df_label['Test Passed'])

print("Encoded DataFrame")
df_label[['Name','Gender','Gender Encoded','Test Passed','Passed Encoded']]
```

# ONE HOT ENCODING 

```python
df_hot=pd.get_dummies(df_label,columns=['City'])
df_hot
```

```python
df_hot=pd.get_dummies(df_label,columns=['City'])
dummy_cols = df_hot.columns.difference(df_label.columns)   # only the new one-hot columns
df_hot[dummy_cols] = df_hot[dummy_cols].astype(int)

df_hot
```
